## 1. üìä Vectores en NumPy

### 1.1 ¬øQu√© es un Vector?
Un vector es un array ordenado de n√∫meros del mismo tipo. En machine learning:
- Representan **features** de un ejemplo
- Almacenan **par√°metros** del modelo
- Contienen **predicciones** y **targets**

```python
# üìç CREACI√ìN DE VECTORES

# Diferentes formas de crear vectors
print("üîπ Diferentes formas de crear vectores:")

# Vector de ceros
a = np.zeros(4)
print(f"Zeros: {a}, shape: {a.shape}, dtype: {a.dtype}")

# Vector de valores aleatorios
b = np.random.random_sample(4)
print(f"Random: {b}, shape: {b.shape}")

# Vector manual
c = np.array([5, 4, 3, 2])
print(f"Manual: {c}, shape: {c.shape}")

# Vector con rango
d = np.arange(4.)
print(f"Arange: {d}, shape: {d.shape}")
```

### 1.3 ‚ö° Operaciones con Vectores

## 2. üéØ Producto Punto (Dot Product)

**El producto punto es FUNDAMENTAL en machine learning:**
- C√°lculo de predicciones: $f_{w,b}(x) = w \cdot x + b$
- Operaciones en redes neuronales
- C√°lculo de similitudes

### F√≥rmula: $a \cdot b = \sum_{i=0}^{n-1} a_i b_i$

```python
# üìç IMPLEMENTACI√ìN MANUAL DEL PRODUCTO PUNTO

def dot_product_manual(a, b):
    """
    Implementa el producto punto manualmente con un loop
    
    Args:
        a (ndarray): vector 1
        b (ndarray): vector 2
    
    Returns:
        resultado (scalar): producto punto
    """
    resultado = 0
    for i in range(len(a)):
        resultado += a[i] * b[i]
    return resultado

# Vectores de ejemplo
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print(f"Vector a: {a}")
print(f"Vector b: {b}")

# Comparar implementaciones
manual_result = dot_product_manual(a, b)
numpy_result = np.dot(a, b)

print(f"\nüîπ Resultados:")
print(f"Producto punto manual: {manual_result}")
print(f"Producto punto NumPy: {numpy_result}")
print(f"¬øSon iguales? {manual_result == numpy_result}")

# Verificaci√≥n paso a paso
print(f"\nüîπ C√°lculo paso a paso:")
print(f"1√ó5 + 2√ó6 + 3√ó7 + 4√ó8 = {1*5} + {2*6} + {3*7} + {4*8} = {1*5 + 2*6 + 3*7 + 4*8}")
```

## 3. ‚ö° La Necesidad de Velocidad: Vectorizaci√≥n vs Loops

**¬øPor qu√© usar NumPy?**
- üöÄ **Velocidad**: 100x-1000x m√°s r√°pido que loops de Python
- üíª **Paralelizaci√≥n**: Aprovecha hardware moderno (SIMD)
- üß† **Memoria**: Uso m√°s eficiente de memoria

```python
# üìç COMPARACI√ìN DE RENDIMIENTO

# Crear vectores muy grandes
print("üîπ Creando vectores de 10 millones de elementos...")
np.random.seed(1)
a_large = np.random.rand(10_000_000)
b_large = np.random.rand(10_000_000)

print(f"Tama√±o de vectores: {len(a_large):,} elementos")

# Prueba con NumPy (vectorizado)
print("\n‚ö° Probando versi√≥n vectorizada (NumPy)...")
start_time = time.time()
resultado_numpy = np.dot(a_large, b_large)
tiempo_numpy = time.time() - start_time

print(f"Resultado NumPy: {resultado_numpy:.4f}")
print(f"Tiempo NumPy: {tiempo_numpy*1000:.2f} ms")

# Prueba con loop manual (solo una muestra peque√±a para no esperar mucho)
print("\nüêå Probando versi√≥n con loop (muestra de 100,000)...")
a_small = a_large[:100_000]
b_small = b_large[:100_000]

start_time = time.time()
resultado_manual = dot_product_manual(a_small, b_small)
tiempo_manual = time.time() - start_time

print(f"Resultado Manual: {resultado_manual:.4f}")
print(f"Tiempo Manual (solo 100K elementos): {tiempo_manual*1000:.2f} ms")

# C√°lculo de speedup
# Extrapolamos el tiempo manual para 10M elementos
tiempo_extrapolado = tiempo_manual * 100  # 10M / 100K = 100
speedup = tiempo_extrapolado / tiempo_numpy

print(f"\nüöÄ An√°lisis de rendimiento:")
print(f"Speedup estimado: {speedup:.0f}x m√°s r√°pido con NumPy")
print(f"Tiempo estimado manual para 10M: {tiempo_extrapolado:.2f} segundos")

# Limpieza de memoria
del(a_large, b_large)
print("\n‚úÖ Memoria liberada")
```

Las matrices son fundamentales en ML:
- **X_train**: Matriz de ejemplos de entrenamiento
- **Cada fila**: Un ejemplo
- **Cada columna**: Una caracter√≠stica (feature)

```python
# üìç CREACI√ìN Y MANIPULACI√ìN DE MATRICES

# Diferentes formas de crear matrices
print("üîπ Creaci√≥n de matrices:")

# Matriz de ceros
matriz_zeros = np.zeros((3, 4))
print(f"Matriz de ceros {matriz_zeros.shape}:")
print(matriz_zeros)

# Matriz manual
print(f"\nMatriz manual:")
X_train = np.array([[2104, 5, 1, 45],    # Casa 1: [tama√±o, cuartos, pisos, edad]
                    [1416, 3, 2, 40],    # Casa 2
                    [852,  2, 1, 35]])   # Casa 3

print(f"X_train shape: {X_train.shape}")
print("X_train:")
print(X_train)

# Informaci√≥n sobre la matriz
print(f"\nüîπ Informaci√≥n de la matriz:")
print(f"Forma (shape): {X_train.shape}")
print(f"N√∫mero de ejemplos (m): {X_train.shape[0]}")
print(f"N√∫mero de features (n): {X_train.shape[1]}")
print(f"Tipo de datos: {X_train.dtype}")
print(f"Tama√±o total: {X_train.size} elementos")
```

```python
# üìç INDEXING EN MATRICES

print("üîπ Acceso a elementos de matrices:")
print(f"X_train:\n{X_train}")

# Acceso a elementos espec√≠ficos
print(f"\nElemento [0,0] (primera casa, tama√±o): {X_train[0, 0]}")
print(f"Elemento [1,2] (segunda casa, pisos): {X_train[1, 2]}")

# Acceso a filas completas (ejemplos)
print(f"\nPrimera casa (fila 0): {X_train[0]}")
print(f"Shape de una fila: {X_train[0].shape}")
print(f"Tipo: {type(X_train[0])}")

# Acceso a columnas completas (features)
print(f"\nTama√±os de todas las casas (columna 0): {X_train[:, 0]}")
print(f"N√∫mero de cuartos (columna 1): {X_train[:, 1]}")

# Slicing m√°s complejo
print(f"\nPrimeras dos casas, primeras dos features:")
print(X_train[:2, :2])

print(f"\n√öltimas dos features de todas las casas:")
print(X_train[:, -2:])
```

## 5. üßÆ Reshape: Cambiando Formas de Arrays

**Reshape** es crucial en ML para:
- Convertir vectores 1D en matrices 2D
- Preparar datos para diferentes algoritmos
- Ajustar dimensiones para operaciones espec√≠ficas

```python
# üìç RESHAPE EN ACCI√ìN

# Vector original
vector_original = np.arange(12)
print(f"Vector original: {vector_original}")
print(f"Shape original: {vector_original.shape}")

# Diferentes reshapes
print(f"\nüîπ Diferentes formas:")

# Matriz 3x4
matriz_3x4 = vector_original.reshape(3, 4)
print(f"\nMatriz 3x4:")
print(matriz_3x4)
print(f"Shape: {matriz_3x4.shape}")

# Matriz 4x3
matriz_4x3 = vector_original.reshape(4, 3)
print(f"\nMatriz 4x3:")
print(matriz_4x3)

# Usando -1 (NumPy calcula autom√°ticamente)
matriz_auto = vector_original.reshape(-1, 2)  # 6x2
print(f"\nMatriz con -1 (auto): {matriz_auto.shape}")
print(matriz_auto)

# Convertir a columna (com√∫n en ML)
columna = vector_original.reshape(-1, 1)
print(f"\nVector columna: {columna.shape}")
print(columna[:5])  # Solo primeros 5 elementos

# Volver a 1D
de_vuelta_1d = matriz_3x4.reshape(-1)  # o .flatten()
print(f"\nDe vuelta a 1D: {de_vuelta_1d}")
print(f"Shape: {de_vuelta_1d.shape}")
```

## 6. üéØ Ejemplo Pr√°ctico: Simulando ML

Veamos c√≥mo todo esto se conecta en un contexto de machine learning:

```python
# üìç EJEMPLO PR√ÅCTICO DE ML

print("üè† Ejemplo: Predicci√≥n de Precios de Casas")
print("=" * 50)

# Dataset de casas (mismos datos de antes)
X_casas = np.array([[2104, 5, 1, 45],    # [tama√±o_sqft, cuartos, pisos, edad]
                    [1416, 3, 2, 40],
                    [852,  2, 1, 35],
                    [1940, 4, 1, 10]])

# Precios reales (en miles)
y_casas = np.array([460, 232, 178, 500])

# Par√°metros del modelo (simulados - como si ya entrenamos)
w = np.array([0.39, 18.75, -53.36, -26.42])  # Pesos para cada feature
b = 785.18  # Bias

print(f"üìä Dataset:")
print(f"X_casas shape: {X_casas.shape} (4 casas, 4 features)")
print(f"y_casas shape: {y_casas.shape} (4 precios)")
print(f"\nFeatures: [tama√±o_sqft, cuartos, pisos, edad]")
print(f"Par√°metros: w = {w}, b = {b}")

# Hacer predicciones para cada casa
print(f"\nüîÆ Predicciones:")
for i in range(X_casas.shape[0]):
    # Extraer features de la casa i
    x_i = X_casas[i]  # Vector de features para casa i
    
    # Predicci√≥n usando producto punto: f(x) = w¬∑x + b
    prediccion = np.dot(x_i, w) + b
    precio_real = y_casas[i]
    
    print(f"Casa {i+1}: {x_i} -> Predicci√≥n: ${prediccion:.1f}k, Real: ${precio_real}k")

# Predicci√≥n vectorizada (todas a la vez)
print(f"\n‚ö° Predicci√≥n vectorizada:")
predicciones_todas = X_casas @ w + b  # @ es equivalente a np.dot para matrices
print(f"Predicciones: {predicciones_todas}")
print(f"Reales:       {y_casas}")
print(f"Errores:      {predicciones_todas - y_casas}")
```

## 7. üìà Visualizaci√≥n de Operaciones Vectoriales

```python
# üìç VISUALIZACI√ìN

# Crear datos para visualizar
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1 + y2  # Suma vectorial

# Crear gr√°fico
plt.figure(figsize=(12, 8))

# Subplot 1: Operaciones b√°sicas
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='sin(x)', linewidth=2)
plt.plot(x, y2, label='cos(x)', linewidth=2)
plt.plot(x, y3, label='sin(x) + cos(x)', linewidth=2, linestyle='--')
plt.title('üîπ Operaciones Vectoriales')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 2: Comparaci√≥n de velocidad
plt.subplot(2, 2, 2)
tama√±os = [1000, 5000, 10000, 50000, 100000]
tiempos_numpy = []
tiempos_loop = []

for tama√±o in tama√±os:
    # Vector de prueba
    a = np.random.rand(tama√±o)
    b = np.random.rand(tama√±o)
    
    # Tiempo NumPy
    start = time.time()
    _ = np.dot(a, b)
    tiempo_np = (time.time() - start) * 1000  # en ms
    tiempos_numpy.append(tiempo_np)
    
    # Tiempo loop (solo para tama√±os peque√±os)
    if tama√±o <= 10000:
        start = time.time()
        _ = dot_product_manual(a, b)
        tiempo_lp = (time.time() - start) * 1000
        tiempos_loop.append(tiempo_lp)
    else:
        tiempos_loop.append(np.nan)

plt.plot(tama√±os, tiempos_numpy, 'o-', label='NumPy (vectorizado)', linewidth=2, markersize=6)
plt.plot(tama√±os[:3], tiempos_loop[:3], 's-', label='Loop manual', linewidth=2, markersize=6)
plt.title('‚ö° Comparaci√≥n de Velocidad')
plt.xlabel('Tama√±o del vector')
plt.ylabel('Tiempo (ms)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.yscale('log')

# Subplot 3: Demostraci√≥n de broadcasting
plt.subplot(2, 2, 3)
vector_base = np.array([1, 2, 3, 4, 5])
escalares = [0.5, 1, 1.5, 2]

for i, escalar in enumerate(escalares):
    resultado = vector_base * escalar
    plt.plot(resultado, 'o-', label=f'vector √ó {escalar}', linewidth=2, markersize=6)

plt.title('üì° Broadcasting: Vector √ó Escalar')
plt.xlabel('√çndice')
plt.ylabel('Valor')
plt.legend()
plt.grid(True, alpha=0.3)

# Subplot 4: Ejemplo de reshape
plt.subplot(2, 2, 4)
data_1d = np.random.randn(20)
data_2d = data_1d.reshape(4, 5)

im = plt.imshow(data_2d, cmap='coolwarm', aspect='auto')
plt.title('üîÑ Reshape: 1D ‚Üí 2D Matrix')
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.colorbar(im, shrink=0.6)

plt.tight_layout()
plt.show()

print("üìä Gr√°ficos generados exitosamente!")
```

```python
# üß™ EJERCICIO 2: Producto Punto Personalizado

print("üß™ Ejercicio 2: Producto Punto Personalizado")
print("=" * 50)

# Dados estos vectores, calcula su producto punto de 3 formas diferentes:
a = np.array([2, 4, 6, 8])
b = np.array([1, 3, 5, 7])

print(f"Vector a: {a}")
print(f"Vector b: {b}")

# M√©todo 1: Loop manual
resultado_1 = 0
for i in range(len(a)):
    resultado_1 += a[i] * b[i]

# M√©todo 2: NumPy dot
resultado_2 = np.dot(a, b)

# M√©todo 3: Sum de multiplicaci√≥n elemento a elemento
resultado_3 = np.sum(a * b)

print(f"\n‚úÖ Resultados:")
print(f"M√©todo 1 (loop):      {resultado_1}")
print(f"M√©todo 2 (np.dot):    {resultado_2}")
print(f"M√©todo 3 (sum(a*b)):  {resultado_3}")
print(f"¬øTodos iguales? {resultado_1 == resultado_2 == resultado_3}")

# Paso a paso
print(f"\nüîç C√°lculo paso a paso:")
for i in range(len(a)):
    print(f"{a[i]} √ó {b[i]} = {a[i] * b[i]}")
print(f"Suma total: {' + '.join([str(a[i] * b[i]) for i in range(len(a))])} = {resultado_1}")
```

### Ejercicio 3: Simulaci√≥n de Dataset ML

```python
# üß™ EJERCICIO 3: Simulaci√≥n de Dataset ML

print("üß™ Ejercicio 3: Simulaci√≥n de Dataset ML")
print("=" * 45)

# Crear un dataset sint√©tico para predicci√≥n de ventas
np.random.seed(42)

# Features: [presupuesto_marketing, empleados, a√±os_experiencia]
n_ejemplos = 6
X_ventas = np.array([
    [10000, 25, 5],   # Empresa 1
    [15000, 30, 8],   # Empresa 2
    [5000,  15, 2],   # Empresa 3
    [20000, 40, 12],  # Empresa 4
    [8000,  20, 4],   # Empresa 5
    [25000, 50, 15]   # Empresa 6
])

# Ventas reales (en miles)
y_ventas = np.array([120, 180, 70, 280, 100, 350])

print(f"üìä Dataset de Ventas:")
print(f"Shape: {X_ventas.shape}")
print(f"Features: [presupuesto_marketing, empleados, a√±os_experiencia]")
print(f"\nDatos:")
for i in range(n_ejemplos):
    print(f"Empresa {i+1}: {X_ventas[i]} -> Ventas: ${y_ventas[i]}k")

# An√°lisis estad√≠stico b√°sico
print(f"\nüìà An√°lisis Estad√≠stico:")
feature_names = ['Presupuesto', 'Empleados', 'Experiencia']

for i, nombre in enumerate(feature_names):
    columna = X_ventas[:, i]
    print(f"{nombre}:")
    print(f"  Media: {np.mean(columna):.1f}")
    print(f"  Min: {np.min(columna)}, Max: {np.max(columna)}")
    print(f"  Std: {np.std(columna):.2f}")

print(f"\nVentas:")
print(f"  Media: {np.mean(y_ventas):.1f}k")
print(f"  Min: {np.min(y_ventas)}k, Max: {np.max(y_ventas)}k")

# Correlaciones simples (usando producto punto normalizado)
print(f"\nüîó Correlaciones con Ventas:")
for i, nombre in enumerate(feature_names):
    feature = X_ventas[:, i]
    # Correlaci√≥n de Pearson simplificada
    corr = np.corrcoef(feature, y_ventas)[0, 1]
    print(f"{nombre}: {corr:.3f}")

print(f"\nüí° Interpretaci√≥n:")
print(f"- Valores cercanos a 1: correlaci√≥n positiva fuerte")
print(f"- Valores cercanos a -1: correlaci√≥n negativa fuerte")
print(f"- Valores cercanos a 0: poca correlaci√≥n")
```
